# Eksploratory Data Analysis (EDA)


```python
df_combined.describe(include="all")
```

```python
print("\nDistribusi Nilai Unik Kolom Kategori:")
for col in ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']:
    value_counts = df_combined[col].value_counts()
    print(f"\nKolom: {col}")
    print(value_counts)
```

```python
#mencari nilai korelasi
print("\nKorelasi Antar Variabel Numerik:")
correlation_matrix = df_combined.select_dtypes(include=['int64', 'float64']).corr()
print(correlation_matrix)
```

```python
# Menghitung rata-rata peminjaman sepeda berdasarkan hari kerja dan hari libur
avg_casual_by_holiday = df_combined.groupby('holiday')['casual'].mean()
avg_registered_by_holiday = df_combined.groupby('holiday')['registered'].mean()
print("\nRata-rata Peminjaman Sepeda Berdasarkan Hari Kerja dan Hari Libur:")
print(avg_casual_by_holiday)
print(avg_registered_by_holiday)


```

```python
# Mengelompokkan data berdasarkan jam dan menghitung rata-rata jumlah pengguna
avg_casual_by_hour = df_combined.groupby('hr')['casual'].mean()
avg_registered_by_hour = df_combined.groupby('hr')['registered'].mean()
print("\nRata-rata Peminjaman Sepeda Berdasarkan Jam:")
print(avg_casual_by_hour)
print(avg_registered_by_hour)

```
